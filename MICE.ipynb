```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Missing values
print(titanic.isnull().sum())
```

```python
data = titanic[['age', 'fare', 'pclass', 'sex']]
data = pd.get_dummies(data, drop_first = True)
data.head()
```

```python
imputer = IterativeImputer(max_iter=10, random_state=0)
```

```python
data_imputed = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)
```

```python
print("Datos originales")
print(data.head(20))
print("\nDatos imputados")
print(data_imputed.head(20))
```

Exercise 1: Use MICE on the life-expectancy dataset https://www.kaggle.com/datasets/kumarajarshi/lif-expectany-whi
Con el dataset completo

```python
import kagglehub
DATASET_PATH = None
# Download the dataset from Kaggle
path = kagglehub.dataset_download("kumarajarshi/life-expectancy-who")

print("Path to dataset files:", path)

life = pd.read_csv(f"{path}/Life Expectancy Data.csv")
life.head()
```

```python
print("Valores faltantes por columna:")
life.isnull().sum()
```

```python
life_numeric = life.select_dtypes(include=[np.number])
life_numeric.head()
```

```python
imputer = IterativeImputer(max_iter=10, random_state=0)

life_imputed = pd.DataFrame(
    imputer.fit_transform(life_numeric),
    columns=life_numeric.columns
)

life_imputed.head()
```
