```python
import seaborn as sns
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Missing values
print(titanic.isnull().sum())
```

```python
data = titanic[['age', 'fare', 'pclass', 'sex']]
data = pd.get_dummies(data, drop_first = True)
data.head()
```

```python
imputer = IterativeImputer(max_iter=10, random_state=0)
```

```python
data_imputed = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)
```

```python
print("Datos originales")
print(data.head(20))
print("\nDatos imputados")
print(data_imputed.head(20))
```

Exercise 1: Use MICE on the life-expectancy dataset https://www.kaggle.com/datasets/kumarajarshi/lif-expectany-whi
Con el dataset completo

```python
import kagglehub
DATASET_PATH = None
# Download the dataset from Kaggle
path = kagglehub.dataset_download("kumarajarshi/life-expectancy-who")

print("Path to dataset files:", path)
```

```python
life = pd.read_csv(f"{path}/Life Expectancy Data.csv")
life.head()
```

```python
# Missing values
print(life.isnull().sum())
```

```python
life.info()
```

```python
# Numeric Columns +  Status
data_life = life.select_dtypes(include=['float64', 'int64']).copy()
data_life = pd.concat([data_life, life['Status']], axis=1)

# One-hot encoding of Status column
data_life = pd.get_dummies(data_life, columns=['Status'], drop_first=True)
data_life.head()
```

```python
# Imputer MICE
imputer = IterativeImputer(max_iter=10, random_state=0)
data_life_imputed = pd.DataFrame(imputer.fit_transform(data_life), columns=data_life.columns)
```

```python
# Original Data - Life Expectancy Dataset
print("Datos originales - Life Expectancy Dataset")
data_life.head(50)
```

```python
# Imputed Data - Life Expectancy Dataset
print("\nDatos imputados - Life Expectancy Dataset")
data_life_imputed.head(50)
```

```python
# Check that there are no missing values in the imputed dataset
print(data_life_imputed.isnull().sum())
```

Exercise 2: Use MICE on the Planets(seaborn) dataset to fill in missing values



```python
# Load dataset
planets = sns.load_dataset('planets')
planets.head()
```

```python
# Missing values
print(planets.isnull().sum())
```

```python
planets.info()
```

```python
# Select only numeric columns
numeric_cols = planets.select_dtypes(include=['float64', 'int64']).columns
data_planets = planets[numeric_cols].copy()
data_planets.head()
```

```python
# Imputer MICE
imputer = IterativeImputer(max_iter=10, random_state=0)
data_planets_imputed = pd.DataFrame(imputer.fit_transform(data_planets), columns=data_planets.columns)
```

```python
print("Datos originales - Planets Dataset")
print(data_planets.head(30))
print("\nDatos imputados - Planets Dataset")
print(data_planets_imputed.head(30))
```

```python
# Check that there are no missing values in the imputed dataset
print(data_planets_imputed.isnull().sum())
```
